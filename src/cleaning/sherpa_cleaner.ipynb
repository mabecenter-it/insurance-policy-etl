## Importal librerias   

```python
import pandas as pd
import numpy as np
import re
```

## Cargar csv

```python
df = pd.read_csv("../../data/raw/marketplace.csv", dtype=str) 
df.head()
```

## FUNCION LIMPIEZA


```python
def clean_text(series):
    return (
        series
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )

text_fields = [
    "first_name", "last_name", "plan_name", "issuer",
    "policy_status", "street_address", "city",
    "preferred_language"
]

df[text_fields] = df[text_fields].apply(clean_text)
```

## NOMBRES CAPITALIZADOS

```python
df["first_name"] = df["first_name"].str.title()
df["last_name"]  = df["last_name"].str.title()
```

## GENDER NORMALIZADO

```python
df["gender"] = (
    df["gender"]
    .str.strip()
    .str.lower()
    .map({
        "m": "Male",
        "male": "Male",
        "f": "Female",
        "female": "Female"
    })
)
```

## TELÉFONO SOLO NÚMEROS Y VALIDAR 10 DIGITOS

```python
df["phone"] = df["phone"].str.replace(r"[^0-9]", "", regex=True)

df["phone"] = df["phone"].where(df["phone"].str.len() == 10)
```

## LIMPIAR EMAIL

```python
df["email"] = df["email"].str.strip().str.lower()

df["email"] = df["email"].where(
    df["email"].str.match(r"^[^@]+@[^@]+\.[^@]+$", na=False)
)
```

## ESTADO POLIZA EN MAYÚSCULAS

```python
df["state"] = df["state"].str.strip().str.upper()
```

## ZIPCODE SOLO 5 NÚMEROS

```python
df["zip_code"] = df["zip_code"].str.replace(r"[^0-9]", "", regex=True)
df["zip_code"] = df["zip_code"].where(df["zip_code"].str.len() == 5)
```

## LIMPIAR FECHAS

```python
date_fields = [
    "effective_date",
    "expiration_date",
    "paid_through_date"
]

for col in date_fields:
    df[col] = pd.to_datetime(df[col], errors="coerce").dt.strftime("%Y-%m-%d")
```

## VALORES MONETARIOS

```python
money_fields = [
    "gross_premium",
    "health_premium",
    "subsidy",
    "net_premium"
]

for col in money_fields:
    df[col] = (
        df[col]
        .str.replace(r"[^0-9.-]", "", regex=True)
        .replace("", np.nan)
        .astype(float)
    )
```

## GUARDAR LIMPIO

```python
df.to_csv("mp_clean_file.csv", index=False)
```
