## Importal librerias   

```python
import pandas as pd
import numpy as np
import re
```

## Cargar csv

```python
df = pd.read_csv("../../data/raw/crm.csv", dtype=str) 
df.head()
```

## LIMPIAR NOMBRES Y APELLIDOS


```python
def clean_name(x):
    if pd.isna(x):
        return None
    x = x.strip()
    x = re.sub(r'\s+', ' ', x)  # quitar espacios dobles
    return x.title()  # Capitalizar

df["first_name"] = df["first_name"].apply(clean_name)
df["last_name"] = df["last_name"].apply(clean_name)
```

## NORMALIZAR GENDER

```python
def clean_gender(x):
    if pd.isna(x):
        return None
    x = x.strip().lower()
    if x in ["m", "male"]:
        return "Male"
    if x in ["f", "female"]:
        return "Female"
    return None

df["gender"] = df["gender"].apply(clean_gender)
```

## LIMPIAR IDS (Sólo números)

```python
def only_numbers(x):
    if pd.isna(x):
        return None
    return re.sub(r'[^0-9]', '', x)

id_fields = ["member_id", "mp_id"]

for col in id_fields:
    df[col] = df[col].apply(only_numbers)
```

## LIMPIAR TELÉFONOS

```python
phone_fields = ["phone1", "phone2", "phone3"]

for col in phone_fields:
    df[col] = df[col].apply(only_numbers)
```

## LIMPIAR EMAIL

```python
df["email1"] = df["email1"].str.strip().str.lower()

# Quitar emails inválidos básicos
df["email1"] = df["email1"].apply(
    lambda x: x if pd.notna(x) and re.match(r"^[^@]+@[^@]+\.[^@]+$", x) else None
)
```

## LIMPIAR FECHAS

```python
date_fields = [
    "efective_date",
    "expiration_date",
    "paid_through"
]

for col in date_fields:
    df[col] = pd.to_datetime(df[col], errors="coerce").dt.strftime("%Y-%m-%d")
```

## LIMPIAR VALORES NUMÉRICOS (PREMIUM, SUBSIDY)

```python
money_fields = ["premium", "police_price", "subsidy"]

def clean_money(x):
    if pd.isna(x):
        return None
    x = re.sub(r'[^0-9.-]', '', x)  # dejar solo números, punto y negativo
    try:
        return float(x)
    except:
        return None

for col in money_fields:
    df[col] = df[col].apply(clean_money)
```

## LIMPIAR DIRECCION

```python
text_fields = ["address", "city", "state", "company", "plan_name", "policy_status"]

for col in text_fields:
    df[col] = df[col].str.strip()

df["state"] = df["state"].str.upper()

df["zipcode"] = df["zipcode"].apply(only_numbers)
df["zipcode"] = df["zipcode"].apply(lambda x: x if x and len(x) == 5 else None)
```

## GUARDAR CSV LIMPIO

```python
df.to_csv("crm_clean_file.csv", index=False)
```
